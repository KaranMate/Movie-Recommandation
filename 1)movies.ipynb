```python
import nltk
from nltk.corpus import stopwords  # Corrected: Added 'from' keyword before nltk.corpus
from nltk.stem import WordNetLemmatizer  # Corrected: Added 'from' keyword before nltk.stem
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocessing_text(text):
    
    #lower case
    text=str(text).lower()
    
    #removed the punctuations
    text = re.sub(r'[^\w\s]', '', text)
    words=text.split()

    #remove stopwords
    words=[i for i in words if i not in stop_words]

    #lemmatize
    words=[lemmatizer.lemmatize(i) for i in words]

    return " ".join(words)
```

```python
 df['tags']=df['tags'].apply(preprocessing_text)
```
